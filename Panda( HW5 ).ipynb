```python
import pandas as pd
```

```python
#Ex1
data = pd.read_csv('Downloads/nyc_weather_2018.csv')
data.head()
```

```python
data1 = pd.read_csv('Downloads/weather_stations.csv')
data1.head()
```

```python
data1.rename(columns = {'id':'station'},inplace = True)
data1.columns
```

```python
weather_snow = data[data.datatype == 'SNOW']
weather_snow.head()
```

```python
station_us = data1[data1.name.str.contains('NJ US')]
station_us.head()
```

```python
info = weather_snow.merge(station_us, on = 'station')
```

```python
info.shape
```

```python
info['station'].unique().shape
```

```python
info.drop_duplicates('station', inplace = True)
info.shape
```

```python
info.elevation.mean()
```

```python
#Ex3
df = pd.read_csv('Downloads/earthquakes.csv')
df.head()
```

```python
df1 = df[df.magType == 'ml']
df = df.assign(bin = pd.cut(df1.mag, bins = 5,labels = ['nor', 'low', 'med', 'high', 'dangerous']))
df.bin.value_counts()
```

```python
#Ex4:
w_station = pd.read_csv('Downloads/weather_by_station.csv')
w_station.head()
```

```python
cal_snow = w_station[w_station.datatype == 'SNOW']
cal_snow.head()
```

```python
cal_snow = cal_snow.assign(month = pd.DatetimeIndex(cal_snow.date).month)
cal_snow.head()
```

```python
cal_snow.groupby(['station', 'month']).mean()
```
